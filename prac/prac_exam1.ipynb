```python
import pandas as pd
import numpy as np

### 00. 데이터 가져오기
car = pd.read_csv('../data/모의고사/01회/Cars93.csv', encoding = 'cp949')

### 01.필요한 칼럼만 가져오기 >> type, Price
car = car[ ['Type', 'Price'] ].copy()
# print(car.info())
# print(car.head())

# print(car['Type'].value_counts())

price = car['Price']
Type = car['Type']
price_large = price[Type == "Large"].reset_index(drop = True).dropna()
price_small = price[Type == "Small"].reset_index(drop = True).dropna()

print('\n-----------------------------결측치')
print(price_large.isnull().sum())
print(price_small.isnull().sum())

### (a) u1 - u2의 점추정량 (소수 둘째자리까지)
print('\n-----------------------------(a)')
u1 = price_large.mean()
u2 = price_small.mean()

result_a = round(u1 - u2, 2)
print("\nresult_a: ", result_a)

### (b) (a)에 대한 표준오차 (소수 둘째 자리까지)
# 검정통계량 = 점추정량/표준오차 >> 표준오차 = 점추정량/검정통계량 >> 표준오차 = result_a/stat
print('\n-----------------------------(b)')
from scipy.stats import ttest_ind, t
stat, pval = ttest_ind(price_large, price_small, equal_var = True)

se = result_a / stat
result_b = round(se, 2)
print(result_b)

### (c) (a)~(b)를 통해 95% 신뢰하한을 구하고 (소수 첫째자리까지), 유의수준 0.05 하에서 가설 검정의 결과 출력
# 자유도
dof = len(price_large) + len(price_small) - 2

# 95% 신뢰하한
CI = t.interval(.95, dof, loc = result_a, scale = se)[0]
# print(CI)

# 신뢰하한 / 기각여부
lower = round(CI, 1)   # >> 이 값이 0보다 크므로 가설은 기각
print(lower)

print('기각')
```
